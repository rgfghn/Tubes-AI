```python
# Node class for decision tree
class Node:
    # konstruktor __init__ menerima beberapa parameter yang memungkinkan untuk
      #  menginisialisasi properti  feature_index, threshold, value,
        #  left, dan right dari objek DecisionTreeClassifier.  
    def __init__(self, feature_index=None, threshold=None, value=None, left=None, right=None): 
        # Menyimpan nilai feature_index, threshold, value, left, right ke 
          # dalam properti masing-masing pada objek DecisionTreeClassifier. 
        self.feature_index = feature_index  # Index of the feature to split on
        self.threshold = threshold  # Threshold value for the split
        self.value = value  # Class label for leaf node
        self.left = left  # Left child (subtree)
        self.right = right  # Right child (subtree)
```

```python
# Decision Tree class
class DecisionTree:
    def __init__(self, max_depth=None):
        # max_depth: Properti ini digunakan untuk menyimpan nilai kedalaman maksimum
          # dari pohon keputusan yang akan dibangun 
        self.max_depth = max_depth  # Maximum depth of the tree
        # root: Properti ini akan digunakan untuk menyimpan referensi ke akar (root)
          # dari pohon keputusan yang akan dibangun. 
        self.root = None  # Root node of the tree

    def fit(self, X, y):
        # kita menginisialisasi properti root dengan memanggil metode 
          # _build_tree dengan argumen X dan y. Metode ini akan mengembalikan 
            # node root dari pohon keputusan yang telah dibangun. 
        self.root = self._build_tree(X, y)

    def _build_tree(self, X, y, depth=0):
        # Check termination conditions
        if depth == self.max_depth or len(np.unique(y)) == 1:
            # Jika kondisi terpenuhi, kita menginisialisasi variabel value dengan 
              # memanggil metode _most_common_label dengan argumen y
            value = self._most_common_label(y)
            # mengembalikan objek node dengan nilai label yang telah ditentukan.
            return Node(value=value)

        # menginisialisasi variabel num_features dengan jumlah fitur dalam data X. X.shape[1] 
          # mengembalikan jumlah kolom dalam matriks X, yang mewakili jumlah fitur.
        num_features = X.shape[1]
        # Variabel best_feature_index diinisialisasi dengan nilai None
        best_feature_index = None
        # Variabel best_threshold diinisialisasi dengan nilai None
        best_threshold = None
        # Variabel best_gini diinisialisasi dengan nilai tak terhingga (inf)
        best_gini = float('inf')

        # Melalui loop ini, kita melakukan iterasi melalui setiap fitur dalam data.
        for feature_index in range(num_features):
            # Baris ini menginisialisasi variabel thresholds dengan array 
              # yang berisi nilai unik dari fitur pada feature_index 
            thresholds = np.unique(X[:, feature_index])
            # Melalui loop ini, kita melakukan iterasi melalui setiap threshold 
              # yang mungkin untuk fitur pada feature_index
            for threshold in thresholds:
                # Baris ini menghitung nilai gini dengan memanggil metode _gini_index() 
                  #  dengan argumen X, y, feature_index, dan threshold
                gini = self._gini_index(X, y, feature_index, threshold)
                # Dalam kondisi ini, kita memeriksa apakah nilai gini yang baru
                  # ditemukan (gini) lebih kecil dari best_gini saat ini. Jika iya,
                    # berarti kita telah menemukan pemisahan yang memiliki 
                      # nilai gini lebih baik, sehingga kita memperbarui best_gini, 
                        # best_feature_index, dan best_threshold dengan nilai yang baru.
                if gini < best_gini:
                    best_gini = gini
                    best_feature_index = feature_index
                    best_threshold = threshold

        # Check termination conditions
        if best_feature_index is None or best_threshold is None:
            # Jika kondisi terpenuhi, kita menginisialisasi variabel value
              # dengan memanggil metode _most_common_label dengan argumen y 
            value = self._most_common_label(y)
            # mengembalikan objek node dengan nilai label yang telah ditentukan.
            return Node(value=value)

        # Baris ini menggunakan indexing untuk memilih kolom dengan 
          # best_feature_index dari matriks X, kemudian membandingkan nilai
            # setiap elemen best_threshold
        left_indices = X[:, best_feature_index] < best_threshold
        # Baris ini menggunakan operator bitwise NOT (~) untuk mengubah nilai 
          # dari array left_indices. Operator ini akan mengubah
            # setiap nilai True menjadi False dan setiap nilai False menjadi True 
        right_indices = ~left_indices

        # Baris ini memanggil metode _build_tree secara rekursif untuk membangun 
          # subtree kiri dari node saat ini, kemudia memanggil depth +1 sebagai argumen
            #  ketiga adalah untuk meningkatkan kedalaman subtree yang sedang dibangun
        left_child = self._build_tree(X[left_indices], y[left_indices], depth + 1)
        # Sama seperti diatas namun baris ini membangun subtree dari kanan
        right_child = self._build_tree(X[right_indices], y[right_indices], depth + 1)

        # Baris ini mengembalikan objek Node yang dibangun dengan menggunakan 
          # informasi terbaik yang ditemukan selama  proses pembangunan pohon.
        return Node(feature_index=best_feature_index, threshold=best_threshold, left=left_child, right=right_child)

    def _gini_index(self, X, y, feature_index, threshold):
        # Baris ini menggunakan indexing untuk memilih kolom dengan 
          # best_feature_index dari matriks X, 
            # kemudian membandingkan nilai setiap elemen dari threshold
        left_indices = X[:, feature_index] < threshold
        # Baris ini menggunakan operator bitwise NOT (~) untuk mengubah 
          # nilai dari array left_indices. Operator ini akan mengubah
            # setiap nilai True menjadi False dan setiap nilai False menjadi True
        right_indices = ~left_indices

        # Baris ini mengambil label/target yang sesuai dengan data yang 
          # masuk ke subtree kiri. Variabel left_labels akan berisi 
           # label/target yang sesuai dengan subset data yang memenuhi 
            # kondisi X[:, feature_index] < threshold
        left_labels = y[left_indices]
        # Sama seperti baris atas namun baris ini mengambil laber/target 
          # yang sesuai dari data yang masuk ke sutree kanan
        right_labels = y[right_indices]

        # Baris ini menghitung gini impurity atau ketidakmurnian gini 
          # dari subset data yang masuk ke subtree kiri. Variabel left_gini 
            # akan menyimpan nilai gini impurity yang dihitung menggunakan 
              # metode _gini_impurity dengan parameter left_labels
        left_gini = self._gini_impurity(left_labels)
        # Baris ini sama seperti baris atas namun bari ini menghitung 
         # ketidakmurnian gini dari subset data yang masuk ke subtree kanan
        right_gini = self._gini_impurity(right_labels)

        # Baris ini menghitung proporsi bobot dari subset data yang masuk ke  
          # subtree kiri. Variabel left_weight akan menyimpan nilai proporsi 
            # tersebut, yang dihitung dengan membagi jumlah sampel dalam  
              # left_labels dengan jumlah total sampel dalam y
        left_weight = len(left_labels) / len(y)
        # Baris ini seperti baris atas namun bari ini menghitung proposisi 
         # bobot dari subset data yang masuk ke subtree kanan
        right_weight = len(right_labels) / len(y)

        # Baris ini menggabungkan nilai gini untuk subtree kiri dan subtree
            # kanan dengan mempertimbangkan bobot relatif masing-masing subtree
              # Dengan memperhatikan bobot ini, nilai gini rata-rata dari kedua 
                # subtree digunakan untuk menghitung gini index yang akhir.
        return left_weight * left_gini + right_weight * right_gini

    def _gini_impurity(self, y):
        # Baris ini digunakan untuk mendapatkan nilai unik dalam array y dan  
          # jumlah Fungsi np.unique dengan argumen return_counts=True kemunculannya.
            # akan mengembalikan dua array, yaitu array nilai unik dan array jumlah
              # kemunculan nilai tersebut. Dalam kasus ini, kita tidak tertarik 
                # dengan array nilai unik, sehingga menggunakan variabel '_'
                  #  untuk menampungnya 
        _, counts = np.unique(y, return_counts=True)
        # Baris ini digunakan untuk menghitung probabilitas kemunculan setiap 
          # kelas dalam array y. Menggunakan counts yang berisi array jumlah 
            # kemunculan nilai unik, dan dibagi dengan panjang y untuk  
              #  mendapatkan probabilitas relatif dari setiap kelas.
        probabilities = counts / len(y) 
        # Baris ini digunakan untuk menghitung indeks gini berdasarkan array
          #  probabilitas yang dihitung sebelumnya.
        gini = 1 - np.sum(probabilities ** 2)
        # Mengembalikan hasil dari perhitungan yang ada dalam 'gini'
        return gini

    def _most_common_label(self, y):
        # Baris ini menggunakan fungsi np.unique dengan argumen return_counts=True 
          # untuk mengembalikan array labels yang berisi nilai-nilai unik 
            # dari y dan array counts yang berisi jumlah kemunculan masing-masing 
              # nilai dalam y. Dengan menggunakan argumen return_counts=True, kita 
                # mendapatkan jumlah kemunculan untuk setiap nilai unik dalam y.
        labels, counts = np.unique(y, return_counts=True)
        # Baris ini mengembalikan label yang paling umum (yang memiliki jumlah 
          # kemunculan terbanyak) dari array labels. Fungsi np.argmax digunakan  
            # untuk mendapatkan indeks elemen maksimum dalam array counts, dan 
              # kemudian indeks tersebut digunakan untuk mengakses elemen yang sesuai 
                # dalam array labels, sehingga mengembalikan label yang paling umum.
        return labels[np.argmax(counts)]

    # Fungsi predict ini mengambil input X, yang merupakan array dari 
      # sampel yang ingin diprediksi
    def predict(self, X):
        # Baris ini menggunakan list comprehension, dilakukan iterasi pada 
          # setiap sampel x dalam X dan dipanggil metode _predict_sample(x)  
            # untuk melakukan prediksi pada setiap sampel tersebut.
        return np.array([self._predict_sample(x) for x in X])

    def _predict_sample(self, x):
        # menginisialisasi variabel node dengan node akar dari pohon keputusan 
        node = self.root
        # loop while yang akan terus berjalan selama nilai node.value adalah None
        while node.value is None:
            # Memeriksa apakah nilai fitur pada sampel x pada indeks yang sesuai 
              # dengan node.feature_index kurang dari node.threshold
            if x[node.feature_index] < node.threshold:
                # Jika kondisi terpenuhi maka akan mengupdate node menjadi node.left,
                  # yaitu mengikuti jalur kiri dari node ke node anak sebelah kiri
                node = node.left
            else:
                # Jika kondisi tidak terpenuhi maka akan mengupdate node menjadi 
                  # node.right, yaitu mengikuti jalur kanan dari node 
                    # ke node anak sebelah kanan
                node = node.right
        # pada baris return node.value menjelaskan bahwa ketika kita mencapai 
          # node daun (leaf node) yang memiliki nilai node.value tidak None, 
            # kita akan mengembalikan nilai tersebut sebagai hasil prediksi.
        return node.value
```

```python
## Hapus kolom terakhir (class) dari dataset ##
# Pada baris tersebut, dilakukan pengambilan nilai-nilai dari dataset df 
  # kecuali kolom terakhir menggunakan metode iloc dari pandas. iloc digunakan 
    # untuk mengindeks dan memilih baris dan kolom berdasarkan posisi numerik.
X = df.iloc[:, :-1].values

## Enkode label kelas (ubah menjadi nilai numerik) ##
# menggunakan metode factorize dari pandas untuk mengubah label kelas menjadi 
  # nilai numerik. factorize digunakan untuk mengindekskan label kelas menjadi nilai numerik
y = pd.factorize(df.iloc[:, -1])[0]

## Bagi dataset menjadi set train dan test ##
# Menghitung ukuran set pelatihan dengan mengalikan 0.8 dengan jumlah baris dalam 
  # dataset X dan mengonversi hasilnya menjadi integer.
train_size = int(0.8 * len(X))
# Memisahkan dataset X menjadi X_train (set pelatihan) dan X_test (set pengujian).
  # X_train akan berisi baris pertama hingga train_size - 1 dari dataset X, 
    # sedangkan X_test akan berisi baris train_size hingga akhir dari dataset X.
X_train, X_test = X[:train_size], X[train_size:]
# Memisahkan label y menjadi y_train (label set pelatihan) dan y_test (label set pengujian).
  #  y_train akan berisi elemen pertama hingga train_size - 1 dari array y, 
    # sedangkan y_test akan berisi elemen train_size hingga akhir dari array y.
y_train, y_test = y[:train_size], y[train_size:]

## Buat dan latih model decision tree ##
#Pada baris kode tersebut, sebuah objek DecisionTree dibuat dengan mengatur max_depth=5, 
  # yang akan menjadi kedalaman maksimum dari pohon keputusan yang akan dibangun. 
tree = DecisionTree(max_depth=5)
# model tree dilatih dengan memanggil metode fit pada objek DecisionTree, dengan 
  # argumen X_train dan y_train. Metode fit akan mengkonstruksi pohon keputusan 
    # berdasarkan data pelatihan yang diberikan.
tree.fit(X_train, y_train)

## Lakukan prediksi pada set pengujian ##
# Metode predict dari objek tree dipanggil dengan argumen X_test, yang merupakan 
  # dataset yang akan diprediksi dan disimpan di y_pred
y_pred = tree.predict(X_test)
```

```python
## Menghitung akurasi ##
# Dalam perhitungan ini, np.sum(y_pred == y_test) menghitung jumlah prediksi yang 
  # sesuai dengan nilai sebenarnya. Kemudian hasilnya dibagi dengan len(y_test) 
    # untuk mendapatkan proporsi prediksi yang benar dari keseluruhan sampel.
accuracy = np.sum(y_pred == y_test) / len(y_test)
# Hasil akurasi tersebut kemudian dicetak menggunakan fungsi print() untuk 
  # memberikan informasi tentang akurasi model.
print(f"Accuracy: {accuracy}")
```
