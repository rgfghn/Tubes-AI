```python
missing_value = dict() # Membuat Kamus Kosong untuk menyimpan informasi yang 
                          # hilang di dalam DataFrame
for column in df: # Digunakan untuk melakukan iterasi setiap kolom dalam DataFrame
    for data in df[column]: # Digunakan untuk melakukan iterasi melalui setiap 
                              # nilai dalam kolom column dalam DataFrame
        if data == '?' or data == '': # Digunakan untuk memeriksa apakah nilai 
                                        # data dalam kolom saat ini sama dengan 
                                          # tanda tanya ('?') atau string kosong ('')
            missing_value[column] = 1 + missing_value.get(column, 0) # Digunakan untuk mengupdate 
                      # dictionary missing_value dengan jumlah nilai yang hilang dalam setiap kolom.
                      
missing_value # Untuk melihat hasil dari dictionary missing_value
```

```python
df = df.drop('J', axis=1) # Menghapus kolom J dari DataFrame
df = df.replace('?', np.NaN) # Digunakan untuk mengganti semua nilai '?' dalam DataFrame
def getcolumn(indx): # Mendefinisikan fungsi bernama getcolumn dengan parameter indx
  for idx,column in enumerate(df.columns): # Melakukan Perulangan dengan fungsi 
                                              # enumerate untuk mengiterasi 
                                                # kolom-kolom dalam DataFrame
      if idx == indx: # Memeriksa apakah indeks sama dengan nilai parameter indx
        return column # Jika sama maka akan mengembalikan nilai column
ind = 0 # Menetapkan variabel ind menjadi nilai 0
for null in df.isnull().sum(): # Melakukan loop for yang menggunakan metode 
            # .isnull().sum() pada DataFrame df. Metode ini menghitung 
              # jumlah nilai null (kosong) dalam setiap kolom DataFrame.

  if null!=0: # jika null tudak sama dengan 0 maka if akan berjalan
    column = getcolumn(ind) # Menggunakan fungsi getcolumn(ind) untuk mendapatkan 
                              # nama kolom berdasarkan indeks ind
    df[column]=df[column].fillna(0) # Pada baris ini nilai Nan akan diganti menjadi 0
    df[column]=pd.to_numeric(df[column]) # Pada baris ini ganti type kolom dari objek ke numerik
    df.loc[df[column]==0,column]= df[column].mean() # Menggunakan df.loc untuk mengakses 
              # nilai-nilai dalam kolom yang memenuhi kondisi df[column]==0. Kemudian, 
                # menggantinya dengan rata-rata dari kolom tersebut menggunakan df[column].mean()
                
    ind+=1 # Menambah nilai dari ind
  else:
    ind+=1 # Jika if tidak terpenuhi maka menambahkan nilai dari ind
```

```python
# Membagi data menjadi beberapa fold untuk proses validasi
fold_1 = (df.iloc[0:150].reset_index(drop = True), 
         df.iloc[150:len(df)].reset_index(drop = True)
        )
fold_2 = (df.iloc[150:301].reset_index(drop = True), 
         pd.concat([df.iloc[0:150], df.iloc[301:len(df)]]).reset_index(drop = True)
        )
fold_3 = (df.iloc[301:len(df)].reset_index(drop = True), 
         df.iloc[0:150].reset_index(drop = True)
        )

test, train = fold_2
train
```

```python
#melakukan normalisasi data dengan min max norm
def norm(df):
  df = (df - df.min()) / (df.max() - df.min())
  return df

#assign fitur dan kelas
X = df.drop('Class', axis = 1) #menghapus kolom Class 
Y = df.Class # Class sebagai kelas

#melakukan normalisasi untuk X (fitur-fitur)
X = norm(X)
X
```

```python
# menghitung jarak menggunakan euclidean distance
def euclidean(x1, x2):
  return np.sqrt(np.sum((x1 - x2)**2))

#Euclidean distance dari data baris pertama dan kedua
euclidean(X.iloc[0], X.iloc[1])
```

```python
#training KNN
def KNN(X_train, y_train, X_test, k): # -k adalah banyaknya ketetanggan / nighbors
  dist = []

  # menghitung distance dari data trainind dan data testing
  for row in range(X_train.shape[0]):
    dist.append(euclidean(X_train.iloc[row], X_test))

  data = X_train.copy()
  data['Dist'] = dist # menambahkan data distance pada data
  data['Class'] = y_train # menambahkan class pada data
  data = data.sort_values(by = 'Dist').reset_index(drop = True) # mengurutkan data berdasarkan

  # mengambil label kelass yang sering muncul
  y_pred = data.iloc[:k].Class.mode()
  return y_pred[0]
```

```python
# Menghitung akurasi dari output berdasar kelas
def acc(y_pred, y_true):
  true = 0
  for i in range(len(y_pred)):
    if y_pred[i] == y_true[i]:
      true += 1
  return true/len(y_pred)
```

```python
# Evaluasi model dengan data fold
def evaluate(fold, k):
  test, train = fold
  X_train, y_train = train.drop('Class', axis = 1), train.Class
  X_test, y_test = test.drop('Class', axis = 1), test.Class
  X_train = norm(X_train)
  X_test = norm(X_test)
  y_preds = []
  for row in range(X_test.shape[0]):
    y_preds.append(KNN(X_train, y_train, X_test.iloc[row], k))

  return (acc(y_preds, y_test))
```

```python
k = 5 # variabel k dengan nilai 5, yang akan digunakan sebagai parameter untuk algoritme KNN.
accs = [] # membuat sebuah daftar kosong accs yang akan digunakan untuk 
            # menyimpan akurasi dari setiap lipatan (fold)
folds = [fold_1, fold_2, fold_3] # memiliki sebuah daftar folds yang berisi lipatan-lipatan
                                    #  data yang akan melakukan evaluasi pada setiap lipatan ini.
for i in range(len(folds)):
  accs.append(evaluate(folds[i], k)) # melakukan evaluasi pada setiap lipatan menggunakan  
                                      # fungsi evaluate. Hasil akurasi dari evaluasi tersebut 
                                        # kemudian ditambahkan ke daftar accs
print(f'Menggunakan k : {k}, dengan rata-rata akurasi : {sum(accs)/3}') # mencetak rata-rata akurasi menggunakan  
                      # rumus sum(accs)/3 (asumsi Anda memiliki 3 lipatan) dan mencantumkan nilai k yang digunakan 
                        # dan rata-rata akurasi dari evaluasi lipatan
```
